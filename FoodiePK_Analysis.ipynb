### 📁 Dataset Columns
| Column | Description |
|--------|-------------|
| `Order_ID` | Unique order identifier |
| `Month` | Month of order (Jan–Dec) |
| `City` | City (Lahore) |
| `Area` | Delivery area (Gulberg, DHA, etc.) |
| `Restaurant` | Restaurant name |
| `Food_Category` | Food type (Biryani, Karahi, etc.) |
| `Item_Name` | Specific menu item |
| `Quantity` | Number of items ordered |
| `Unit_Price` | Price per item (PKR) |
| `Discount` | Discount applied (0–20%) |
| `Delivery_Fee` | Delivery charges (PKR) |
| `Subtotal` | Price × Quantity |
| `Total_Amount` | Final amount after discount + delivery |
| `Cost` | Restaurant's cost |
| `Profit` | Net profit per order |
| `Meal_Time` | Breakfast / Lunch / Dinner / Late Night |
| `Order_Status` | Delivered / Cancelled |
| `Payment_Method` | JazzCash / EasyPaisa / COD / Card |
| `Rating` | Customer rating (3.0–5.0) |

## 📦 Step 1 — Importing Libraries & Loading Data

```python
import pandas as pd
import numpy as np

df = pd.read_csv('lahore_foodie_data.csv')
df.head()
```

## 🔍 Step 2 — Data Exploration

```python
# Dataset shape — rows & columns
df.shape
```

```python
# Column names and data types
df.dtypes
```

```python
# Missing values check
df.isnull().sum()
```

```python
# Note: Rating has 400 nulls — these belong to Cancelled orders (no rating given)
# Duplicate rows check
df.duplicated().sum()
```

## 📊 Step 3 — Basic KPIs (Key Performance Indicators)

```python
# Total number of orders
print('Total Orders     :', len(df))
```

```python
# Total sales revenue
print('Total Sales (PKR):', df['Total_Amount'].sum())
```

```python
# Total profit
print('Total Profit (PKR):', df['Profit'].sum())
```

```python
# Average order value
print('Avg Order Value  :', round(df['Total_Amount'].mean(), 2))
```

```python
# Overall profit margin %
print('Profit Margin %  :', round(df['Profit'].sum() / df['Total_Amount'].sum() * 100, 2))
```

## 🍕 Step 4 — Food Category Analysis

```python
# Orders per category
df['Food_Category'].value_counts()
```

```python
# Profit per category
df.groupby('Food_Category')['Profit'].sum().sort_values(ascending=False)
```

```python
# Highest profit category
df.groupby('Food_Category')['Profit'].sum().nlargest(1)
```

```python
# Lowest profit category
df.groupby('Food_Category')['Profit'].sum().nsmallest(1)
```

```python
# Average order value per category
df.groupby('Food_Category')['Total_Amount'].mean().sort_values(ascending=False)
```

## 🏪 Step 5 — Restaurant Analysis

```python
# Top 5 restaurants by Total Sales
df.groupby('Restaurant')['Total_Amount'].sum().nlargest(5)
```

```python
# Top 5 restaurants by Profit
df.groupby('Restaurant')['Profit'].sum().nlargest(5)
```

```python
# Top 5 restaurants by Average Rating
df.groupby('Restaurant')['Rating'].mean().nlargest(5)
```

## 📅 Step 6 — Monthly Trend Analysis

```python
# Orders per month
df['Month'].value_counts()
```

```python
# Best month by Sales
df.groupby('Month')['Total_Amount'].sum().nlargest(1)
```

```python
# Monthly Profit — sorted low to high
df.groupby('Month')['Profit'].sum().sort_values()
```

## 🌙 Step 7 — Meal Time Analysis

```python
# Orders by meal time
df['Meal_Time'].value_counts()
```

```python
# Late Night — most ordered food categories
df[df['Meal_Time'] == 'Late Night']['Food_Category'].value_counts()
```

## 💳 Step 8 — Payment Method Analysis

```python
# Most popular payment methods
df['Payment_Method'].value_counts()
```

```python
# Profit by payment method
df.groupby('Payment_Method')['Profit'].sum().nlargest(5)
```

## ❌ Step 9 — Cancelled Orders Analysis

```python
# Delivery vs Cancellation rate %
df['Order_Status'].value_counts() / len(df) * 100
```

```python
# Cancellations by Food Category
df[df['Order_Status'] == 'Cancelled']['Food_Category'].value_counts()
```

```python
# Cancellations by Area
df[df['Order_Status'] == 'Cancelled']['Area'].value_counts()
```

## 🏷️ Step 10 — Discount Impact Analysis

```python
# Correlation: Does more discount = less profit?
print('Discount vs Profit Correlation:', round(df['Discount'].corr(df['Profit']), 4))
print('Result: Negative correlation — more discount = less profit')
```

```python
# Discount usage distribution
df['Discount'].value_counts()
```

```python
# Average profit when discount > 15%
print('Avg Profit (Discount > 15%):', round(df[df['Discount'] > 0.15]['Profit'].mean(), 2))
```

## ⭐ Step 11 — Rating Analysis

```python
# Top 5 restaurants by Rating
df.groupby('Restaurant')['Rating'].mean().nlargest(5)
```

```python
# Bottom 5 restaurants by Rating
df.groupby('Restaurant')['Rating'].mean().nsmallest(5)
```

```python
# Correlation: Does higher rating = more profit?
print('Rating vs Profit Correlation:', round(df['Rating'].corr(df['Profit']), 4))
```

## 🏘️ Step 12 — Area-wise Performance

```python
# Orders per area
df['Area'].value_counts()
```

```python
# Top 5 most profitable areas
df.groupby('Area')['Profit'].sum().nlargest(5)
```

```python
# Top 5 areas by Average Order Value
df.groupby('Area')['Total_Amount'].mean().nlargest(5)
```
